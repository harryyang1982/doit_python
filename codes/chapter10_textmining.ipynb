```python
moon = open('../datasets/speech_moon.txt', encoding='UTF-8').read()
moon
```

```python
import re 
moon = re.sub('[^가-힣]', ' ', moon)
moon
```

```python
import konlpy
```

```python
hannanum = konlpy.tag.Hannanum()

hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
nouns = hannanum.nouns(moon)
nouns
```

```python
import pandas as pd
df_word = pd.DataFrame({'word': nouns})
df_word
```

```python
df_word = df_word.assign(count = df_word.word.str.len())
```

```python
df_word
```

```python
df_word = df_word.query('count >= 2')
df_word.sort_values('count', ascending=False)
```

```python
df_word = df_word.groupby('word', as_index = False)\
    .agg(n = ('word', 'count'))\
        .sort_values('n', ascending=False)
df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update({'font.family':'AppleGothic',
                     'figure.dpi' : 120,
                     'figure.figsize' : [6.5, 6]})
plt.rcParams['axes.unicode_minus'] = False
```

```python
sns.barplot(data = top20, y = 'word', x = 'n')
```

## 10-2 기사 댓글 텍스트 마이닝

```python
df = pd.read_csv("https://github.com/youngwoos/Doit_Python/raw/main/Data/news_comment_BTS.csv",
                 encoding = 'UTF-8')

df.info()
```

```python
df.reply = df.reply.str.replace('[^가-힣]', ' ', regex = True)
```

```python
df.reply.head()
```

```python
import konlpy
```

```python
kkma = konlpy.tag.Kkma()
```

```python
nouns = df.reply.apply(kkma.nouns)
nouns
```

```python
nouns = nouns.explode()
nouns
```

```python
df_word = pd.DataFrame({'word': nouns})
```

```python
df_word['count'] = df_word.word.str.len()
```

```python
df_word
```

```python
df_word = df_word.query('count >= 2')
```

```python
df_word
```

```python
df_word = df_word.groupby('word', as_index = False)\
    .agg(n = ('word', 'count')) \
        .sort_values('n', ascending = False)
df_word
```

```python
top20 = df_word.head(20)
```

```python
top20
```

```python
plt.rcParams.update({'figure.figsize': [6.5, 6]})

sns.barplot(data = top20, y = 'word', x = 'n')
```
