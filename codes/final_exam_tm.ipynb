```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import konlpy
```

```python
hannanum = konlpy.tag.Hannanum()
```

```python
yoon = open('datasets/윤석열_연설문.txt', encoding = 'UTF-8').read()
```

```python
import re
```

```python
yoon = re.sub('[^ㄱ가-힣]', ' ', yoon)
yoon
```

```python
nouns = hannanum.nouns(yoon)
nouns
```

```python
mecab = konlpy.tag.Mecab()
```

```python
nouns = mecab.nouns(yoon)
```

```python
nouns
```

```python
kkma = konlpy.tag.Kkma()
```

```python
nouns2 = kkma.nouns(yoon)
```

```python
df_word = pd.DataFrame({'word': nouns})
df_word.head()
```

```python
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
df_word = df_word.query('count >= 2')
df_word.sort_values('count', ascending = False)
```

```python
df_word = df_word.groupby('word', as_index=False) \
        .agg(n = ('word', 'count')) \
            .sort_values('n', ascending = False)
```

```python
df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
plt.rcParams.update({'font.family' : 'AppleGothic',
                    'figure.dpi': '120',
                    'figure.figsize': [6.5, 6]})
```

```python
sns.barplot(data = top20, y = 'word', x = 'n')
```

```python
font = 'BMDOHYEON_otf.otf'
```

```python
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
from wordcloud import WordCloud
wc = WordCloud(random_state = 1234,
               font_path = font,
               width = 400,
               height = 400,
               background_color = 'white')
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)
```

```python
plt.figure(figsize = (10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
```
