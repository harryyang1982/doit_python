```python
midwest = pd.read_csv('https://github.com/tidyverse/ggplot2/raw/main/data-raw/midwest.csv')
```

```python
midwest.describe()
```

```python
midwest.info()
```

```python
midwest=midwest.rename(columns={'poptotal':'total', 'popasian':'asian'})
```

```python
midwest['asia_per_total'] = midwest['asian'] / midwest['total'] * 100
```

```python
midwest.asia_per_total.plot.hist()
```

```python
midwest
```

```python
criteria = midwest.asia_per_total.mean()
```

```python
midwest['perc'] = np.where(midwest['asia_per_total'] >= criteria, 'large', 'small')
```

```python
midwest
```

```python
midwest.perc.value_counts().plot.bar()
```
