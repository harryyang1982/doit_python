```python
import pandas as pd
import numpy as np
import pydataset 
from statsmodels.formula.api import ols
```

```python
mtcars = pydataset.data('mtcars')
fifa22 = pd.read_csv('datasets/players_fifa22.csv')
```

```python
mtcars.corr() ** 2 # 제곱하면 결정계수, 제곱안하면 상관계수
```

```python
mod = ols('mpg ~ wt', data = mtcars).fit()
```

```python
mod
```

```python
mod.summary()
```

```python
# mpg = 37.2851 -5.3445 * wt
```
